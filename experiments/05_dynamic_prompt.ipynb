### Prompt Template - Dynamin Prompy

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

 from langchain_core.output_parsers import StrOutputParser  ishka use humne isha liye kiya hai taki hame bar bar result.conternt na likhna pade bas hume har bar output only string ke form me mile

```python
llm = ChatOpenAI(model="gpt-4o")
out = StrOutputParser()
```

```python
prompt = ChatPromptTemplate.from_messages([
    {"role":"system","Content":"You are a translater and tranlate input in {language}"},
    {"role":"user","Content":"{query}"},
])
```

```python
final_prompt = prompt.format_messages(language= "Hindi", query = ""I Love python and JS)
result = llm.invoke(final_prompt)
print(result.content)
```

#### Chains - Runnable

```python
result1 = llm.invoke(prompt.invoke({"language":"Hindi", "query":"I LOVE YOU"}))
print(result1.content)
```

#### Chains


```python
chains = prompt | llm | out
```

langchain ka use hum ishi liye karte hai ki upper wala itna code na likha kar sidha sidha ye likha de taki llm ka output prompt me chala jaye       

```python
result3 = chains.invoke({"language":"Hindi", "query":"I LOVE YOU"})
print(result3)
```
