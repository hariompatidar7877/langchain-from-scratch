### Solutiob is : Structured Output

```python
from pydantic import BaseModel, Field
from typing import List
```

```python
class ResponseStructure(BaseModel)
    name:str = Field(description="Full name")
    gmail:str = Field(description="Email")
    age:int = Field(description="age")



Structure_llm = llm.with_structured_output(ResponseStructure)
result = Structure_llm.invoke(f"Please give me only name,gmail and age from text {text}")
```
