### History but not standered format

### Mange History

```python
history = []
while True:

    query = input("user:  ")

    if query.lower() in ["exit","bye","quit"]:
        print("Good Bye")
        break

    history.append({"role":"user", "content": query})
    print("user:" query)

    res = llm.invoke(query)
    history.append({"role":"user", "content": res})
    print("user:" res)
    
```
